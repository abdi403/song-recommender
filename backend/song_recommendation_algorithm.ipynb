```python
import pandas as pd
import sqlite3
import os
```

```python
root_dir = os.environ.get("PROJECT_DIR")
```

```python
root_dir
```

```python
def get_connection():
    conn=sqlite3.connect(f"{root_dir}/database/song_recommendations_db.db")
    conn.row_factory = sqlite3.Row
    return conn
```

```python
def all_songs():
    conn=get_connection()
    cursor=conn.cursor()
    cursor.execute("SELECT * FROM songs")
    songs= []

    song_objects = cursor.fetchall()
    conn.close()

    for s in song_objects:
        songs.append(dict(s))

    return songs
```

```python
data = all_songs()
```

```python
len(data)
```

```python
songs_df = pd.DataFrame(data)
```

```python
songs_df.head()
```

```python
num_feat= songs_df.select_dtypes(include='number').columns.to_list()
```

```python
x = songs_df[num_feat]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test = train_test_split(x, test_size =0.2, random_state =42)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors = 10, metric='cosine', algorithm='brute') #Building the model
```

```python
model.fit(x_train)
```

```python
x_test.sample()
```
